```python
import nltk
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
data = pd.read_csv("C:/Users/harsh/Desktop/project/Abolfazl Safikhani.csv", header=0,
                    delimiter=",")
df= data['descriptions']
```

```python
df
```

```python
#Breaking the text into sentences
nltk.download('punkt')

sentences = []
for item in df:
	sentence = nltk.sent_tokenize(item)
	sentences.append(sentence)
```

```python
print(sentences)
```

```python
print(len(sentences))
```

```python
for i in range(len(sentences)):
	for j in sentences[i]:
		print(j)
```

```python
# cleaning the text

corpus = []

for i in range(len(sentences)):
	for j in sentences[i]:
		review = re.sub('[^a-zA-Z]',' ',j)
		review = review.lower()
		corpus.append(review)


```

```python
print(corpus)
```

```python
#initializing stemming and lemmatizing

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
 
```

```python
bag_of_words = []
for i in corpus :
	words = nltk.word_tokenize(i)
	for word in words:
		if word not in set(stopwords.words('english')):
			token = stemmer.stem(word)
			token = lemmatizer.lemmatize(token)
			bag_of_words.append(token)
```
